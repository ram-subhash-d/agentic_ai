# Variables and data types
Data type is automatically assigned based on the value. Some of the well known data structures are
* List : An ordered, mutable collection of items enclosed in square brackets []. It can store heterogeneous data types.
* Tuple : An ordered, immutable collection of items enclosed in parentheses (). It can store heterogeneous data types.
* Set : An unordered collection of unique items enclosed in curly braces {}.
* Dictionary : A collection of key-value pairs enclosed in curly braces {}, where each key is unique and immutable, and values can be of any data type and are accessed using their keys.

```python
a = 10 # a is an integer
b = 6.2 # b is a float
c = "Ram" # c is a string
d = True # d is boolean 
e = [1,2,3,4,5] # e is a list
f = (1,2,3,4,5) # f is a tuple
g = {1,2,3,4,5} # g is a set

type(b) # type function will give the type of the variable
```

# List

```python
fruits = ["apple", "bananna", "cherry"]
print(fruits)
print(fruits[0]) # Output is ordered, hence can be accessed by index

fruits.append("date") # Add an item
fruits.remove("bananna") # Remove an item
fruits[1] = "blueberry" # Change an item, use negative index to index from the 
                        # end of the list, -1 is the last element, -2 is the
                        # last but one element, etc. 
print(fruits)

#len function can give the number of items in any data structure
```

# Tuple

```python
fruits = ("apple", "bananna", "cherry")
print(fruits)
print(fruits[0]) # Output is ordered, hence can be accessed by index
# Positive and negetive indexes are supported like list
# We cannot add, remove or change an item in tuples
```

# Set

```python
fruits = {"apple", "bananna", "cherry", "apple"}
print(fruits) # Unordered and unique, duplicate elements are 
# removed, cannot be indexed

fruits.add("date") # Add an item
fruits.remove("bananna") # Remove an item
print(fruits)

#Python has operations to do union and intersection on sets
#Sets cannot contain mutable elements like lists, dictionaries or sets
#In python basic data types like int, float, string are also immutable
```

# Dictionary

```python
student_ages = {"Alice" : 25, "Bob" : 22, "Charlie" : 23, "Alice" : 26}
print(student_ages) # Duplicate keys are overriden
print(student_ages["Alice"]) # Access values by key
student_ages["David"] = 24 # Add a new key value pair, key not already present
student_ages["Alice"] = 27 # Modify an existing key value pair
print(student_ages)
print(student_ages.get("Don", "Unknown")) # Access a value by key, return
                                          # the given value if not found

#Keys should be immutable, values can be anything
user = {"name" : "John Doe",
        "age" : 30} # heterogeneous example
```

# Strings

```python
str = "hello, world! " # Strings are ordered immutable sequence of characters
                       # Can use " or '
# Use """ for multi line strings 
multi_line  = """
Line 1
Line 2
"""
print(multi_line)

# Some string methods
print(str.lower())
print(str.upper())
print(str.capitalize())
print(str.title())
print(str.strip()) # Removes leading and trailing whitespaces
print(str.lstrip()) # Removes trailing whitespaces

print(str.split()) # Splits the string into a list of substrings based
                   # on a specified delimiter
print(" ".join(["hello", "world"])) # Join the list of strings into a single
                                    # string with a specified delimiter
                                    # Delimiter is the starting string
                                    # space in this case
#Python has no character type, so in the above example if you give a delimiter,
#you have to give it as a string
```

```python
# String concatenation
str1 = "hello"
str2 = "world"
str3 = str1 + " " + str2
print(str3)

# String formating
name = "Alice"
age = 30
f_string = f"My name is {name} and I am {age} years old."
print(f_string)
f_string = "My name is {} and I am {} years old.".format(name, age)
print(f_string)
```

# Conditionals : if-else statements
Python uses indentation to delimit blocks of code. Indentation is used to define scope and structure of loops, conditionals, functions, classes and other control structures.

```python
x = 3
if x > 5:
    print("x is greater than 5")
elif x > 2:
    print("x is greater than 2 but less than or equal to 5")
else:
    print("x is 2 or less")
```

# Loops : for and while
* for : Iterate over a sequence
* while : Execute a block based on condition
It has 3 loop controls break and continue.

```python
fruits = ["apple", "bananna", "cherry"]
for fruit in fruits:
    print(fruit)

for i in range(len(fruits)):
    print(fruits[i])

count = 1
while count <= 3:
    print(count)
    count += 1

# range(5) gives the numbers from 0 to 4
```

# Functions
Defined using the def keyword. They are first class citizens.

```python
def greet(name, message="Hello"):
    print(f"{message}, {name}!")

greet("Bob", "Hi") # Positional arguments
greet(message="Hi", name="bob") # Keyword arguments
greet("Alice") # Use default parameter value 

text = "Python is versatile"
def split_text(text):
    words = text.split()
    return words

words = split_text(text)
print(words)

#Allows a function to accept any number of args in a tuple
def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

print(multiply(1, 2, 3, 4))

#Allows a function to accept any number of key-word args in a 
# dictionary
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key} : {value}")

print_info(name="Alice", age=30, city="New York") # Like function call with
                                                  # keyword arguments

# We can specify parameter and return types for a function
# But they are there only to help while calling the function
# Python is not a statically typed language
# For example the below function can be called with floats too
def add_two_numbers(num_1 : int, num_2 : int) -> int:
    return num_1 + num_2

#Lambda functions
lambda x : len(x) 
```

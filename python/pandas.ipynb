Pandas is powerful python library widely used for data manipulation and analysis. It has data structures and functions that simplify working with structured data(tables). It uses NumPy internally. Below are some of the things we can do with Pandas.
* Efficient data handling : Excel cannot handle millons of records in tables, it becomes very slow, Pandas can do it efficiently.
* Data alignment
* Handling missing data
* Data integration
* Flexible data transformation
* Integration with other libraries  

Pandas library runs on a single thread, but still very efficient.  
Pandas is sometimes used as a tool in agents to do mathematical computations, to enhance performance. LLM's are bad at doing mathematical computations.

# Data structures
* Series : One dimentional array with axis labels, like a column in excel.
* DataFrame : Two dimentional data structure with labeled rows and columns, like a sheet in excel(table).

Pandas can also be used to read non-relational databases, by loading the file items into a single column.

```python
# Create a dataframe by reading data from a csv file
csv_url = "data/train.csv"
df = pd.read_csv(csv_url) # Pandas has many functions to read from different file types(data sources) 
                          # like read_json to read from json files
                          # It can also read from databases, like read_sql
df.head(5) # Read the first 5 rows
```

```python
df.info() # Get some basic information about the dataset
df.describe() # Get statistics for all the numeric columns
```

```python
# Select a single column, returns a sereis
df["Age"]
# Select a group of columns, returns a dataframe
df[["Age", "Fare"]]
```

```python
# Filtering, similar to NumPy
df["Fare"] > 15 # Returns a True/False column with which we can filter
df[df["Fare"] > 15] # Records who paid more than 15 pounds
df[df["Fare"] > 15]["Fare"] # Get the Fare column for the records who paid more than 15 pounds
df[df["Fare"] > 15]["Fare"].mean() # Get the average fare of people paying more than 15 pounds
```

```python
# Sorting
df.sort_values("Age", ascending=False)
```

```python
# For an empty cell value, pandas fills it up with NaN(Not a number)
# Replacing an empty value
df.dropna() # Drop all records with null in any of the columns and return the new df
df["Age"].isna() # Returns true for all the rows which have NaN for Age column, else false
df["Age"].isna().sum() # Count the number of rows with NaN values
df["Age"] = df["Age"].fillna(df["Age"].mean()) # Fillup the NaN's of the age column with the average value
```

```python
# Data analysis : groupby, aggregate
# Does sex, age or pclass has anything to do with thier survival rate
df.groupby("Pclass").agg({"Survived" : "mean"})
df.groupby("Sex").agg({"Survived" : "mean"})
df.groupby(["Pclass", "Sex"]).agg({"Survived" : "mean"})
df.groupby("Survived").agg({"Fare" : "mean", "Age" : ["mean", "median"]}) # Fare and age based on survival
```

```python
# Get the value based on index, iloc and loc
df.iloc[0,0] # Get the value based on row index and column index
df.loc[0, "Survived"] # Get the value based on row index and column name

# Drops rows and columns, axis = 0 stands for rows and axis = 1 stands for columns
df.drop("PassengerId", axis=1)
```

```python
# Apply, to apply a custom function to a column
# Add a new column in titanic which tells if the person is eligible to vote
df["eligible_to_vote"] = df["Age"].apply(lambda x: x > 18)
df.head()
```

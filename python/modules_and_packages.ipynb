# Module
A module is a python file containing python code. It can define functions, classes and variables. Use it to organize your code into reusable pieces. It is like a library.

```python
#mymodule.py
def greet(name):
    return f"Hello, {name}!"

def add(a, b):
    return a + b

#main.py
import mymodule

print(mymodule.add(3,4))
```

# Package
A package is a collection of modules organized in directories that includes a special __init__.py file.  
mypackage/  
-- __init__.py  
-- module1.py  
-- module2.py  

```python
from mypackage import module1, module2

print(module1.function1())
print(module2.function2())
```
pip is the package installer for python.  
Library is also used interchangeably with package.

# Virtual environment
A virtual environment is an isolated python environment that allows you to manage dependencies for your projects independently of each other. If you not use a virtual environment, all the installed packages will be installed globally, potentially causing conflicts with other projects and cluttering your global python environment.  

python -m venv projectA_env # Creating a new virtual environment  
python -m venv projectB_env

projectA_env\Scripts\activate # switches your current session to 
                              # use projectA_env virtual environment

(projectA_env) pip install requests==2.24.0 # Install requests package
                                            # into projectA_env
